```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data into a pandas dataframe
data = {'support': [0.03, 0.025, 0.02],
        'total_selected_rules_for_hiding': [10, 15, 100],
        'shapes_before': ['(150, 1486)', '(150, 1486)', '(150, 1486)'],
        'shapes_after': ['(150, 1485)', '(150, 1483)', '(150, 1471)'],
        'no_of_rules_before_hidden': [39, 106, 1072],
        'no_of_rules_after_hidden': [16, 57, 475],
        'rules_lost_ratio': [41.02564102564102, 53.77358490566038, 44.309701492537314],
        'difference_in_total_items_in_each_row': [63, 100, 148],
        'utility_preservation': [0.9800443458980045, 0.968324358568261, 0.9531200506810262],
        'difference_in_rules': [23, 49, 597],
        'hidden_rules_present_after_hiding': [0, 0, 0]}
df = pd.DataFrame(data)

# Bar chart for rules before and after hiding
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['no_of_rules_before_hidden'], label='Before Hiding', alpha=0.5)
plt.bar(df.index, df['no_of_rules_after_hidden'], label='After Hiding', alpha=0.5)
plt.xlabel('Support Value')
plt.ylabel('Number of Rules')
plt.title('Rules Before and After Hiding')
plt.legend()
plt.show()

# Line chart for rules lost ratio
plt.figure(figsize=(10, 6))
plt.plot(df['support'], df['rules_lost_ratio'])
plt.xlabel('Support')
plt.ylabel('Rules Lost Ratio')
plt.title('Rules Lost Ratio vs Support')
plt.show()

# Scatter plot for utility preservation
plt.figure(figsize=(10, 6))
plt.scatter(df['support'], df['utility_preservation'])
plt.xlabel('Support')
plt.ylabel('Utility Preservation')
plt.title('Utility Preservation vs Support')
plt.show()

# Pie chart for total selected rules for hiding
plt.figure(figsize=(10, 6))
plt.pie(df['total_selected_rules_for_hiding'], labels=df['support'], autopct='%1.1f%%')
plt.axis('equal')
plt.title('Distribution of Total Selected Rules for Hiding')
plt.show()

# Heatmap for correlation
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Histogram for difference in total items in each row
plt.figure(figsize=(10, 6))
plt.hist(df['difference_in_total_items_in_each_row'], bins=10, edgecolor='black')
plt.xlabel('Difference in Total Items in Each Row')
plt.ylabel('Frequency')
plt.title('Distribution of Difference in Total Items in Each Row')
plt.show()
```

```python
# Scatter plot for support vs. utility preservation
df.plot(kind='scatter', x='support', y='utility_preservation')
plt.title('Support vs. Utility Preservation')
plt.ylabel('Utility Preservation')
plt.xlabel('Support')
plt.show()
```
